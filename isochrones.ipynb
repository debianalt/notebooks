```python
# Import libraries
import ee
import geemap

# Initialize the Earth Engine API
ee.Initialize()

# Create an interactive map
Map = geemap.Map()
Map.centerObject(ee_object=ee.Geometry.Point([-8.15, 41.55]), zoom=8)

# Define a Point of Interest (POI)
poi = ee.Geometry.Point([-8.15, 41.55])

# Load an elevation dataset
elevation = ee.Image('USGS/SRTMGL1_003')

# Set parameters for the cost distance transformation
max_distance = 50000  # Maximum distance in meters
cell_resolution = 1000  # Resolution in meters

# Create a cost image where the value of each pixel is the elevation
cost_image = elevation

# Perform the cost distance transformation
cumulative_cost = cost_image.cumulativeCost(
    source=poi,
    maxDistance=max_distance,
    geodeticDistance=True
)

# Define isochrone intervals in meters
isochrone_intervals = [10000, 20000, 30000, 40000, 50000]

# Create isochrone image
isochrones = cumulative_cost.expression(
    'interval', {
        'interval': ee.Image.constant(isochrone_intervals)
    }).gt(cumulative_cost)

# Add layers to the map
Map.addLayer(isochrones, {'min': 0, 'max': 1, 'palette': ['FF0000', '00FF00']}, 'Isochrones')

# Show map
Map
```
